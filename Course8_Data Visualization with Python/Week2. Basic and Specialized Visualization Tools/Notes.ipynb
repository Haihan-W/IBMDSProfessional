# All the following Plots: Area Plot, Bar Plot, Histogram Plot, Pie Chart, ... are all built-in Pandas Dataframe's plot method: df.plot(kind='xxx',...)

## 1. Area Plot

- Area Plot shows cumulated totals using numbers or percentages over time
- The default area plot's stack is set to True, in this case, the value of one country is stack over another.
- For example, in the graph below: For Y-Axis: in 1980, India is 8880 (in red), and China is 5123 (show as difference between blue and red color's y value), and UK is 22045 (Purple minus Blue), and so on. Each country's immigrant number is added to previous country's and show as its own color. The final country's y value (Pakistan) here shows the total immigrant number of all 5 countries. (This is only when stack is set to True (default))

## 2. Histograms

- Definition: Represent the FREQUENCY distribution of A VARIABLE (e.g. # of immigrants in example below), it is different from a Bar Chart in next section!!!
- Note 1: pyplot's histogram default bin = 10 (i.e. split x axis range to 10 bins - i.e. 10 bars)
- Note 2: np.histogram() -> purpose is to define the x-tick's interval, the default bin is also 10. So that the x-tick matches the histogram's bin as above, which will visually look better.

- Definition: Unlike a Histogram, a Bar Chart is commonly used to compare the value of a Variable at a given point in time.
