# Content-based Recommendation System (Content-based Filtering)

- Step 1: Create User Profile.
    - For example: User watched different types of movies and have different rating for different types, then build matrix to reflect user preference of movie types.
- Step 2: Recommendation to User based on user profile.
    - For example: There are several new movies to recommend to user. But before that, using user's profile to determine which movie user is mostly likely giving high rating, and recommend that movie.

# Collaborative Filtering

Two types:
- 1. User-based:
    - Two Users are considered as similar user if they both watched similar movies.
    - Then if one user watch a new movie, the other user will get recommended for the same.
    
- 2. Item-based:
    - Two movies are considered as similar movies if they are both watched by same people (e.g. user 1 and user 2 in screenshot)
    - Please note: it is different from Content-based recommendation system, here we don't use any of the move types/attributes (e.g. sci-fi, adventure, action, etc.) or build any user profile. Here we purely based on user behaviour (e.g. movies watched/not watched by same group of people) to group movies.
    - The if a new user (user 3) watched one of the two movies above, then the other movie will be recommended to this new user.
