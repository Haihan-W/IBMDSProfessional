### Watson Knowledge Catalog

Watson Knowledge Catalog unites all information assets into a single metadata-rich catalog, based on Watsons understanding of relationships between assets and how theyre being used and socialized among users in existing projects

This tool covers the following DS tasks (shown on green):
- Data Asset Management
- Data Management
- Data Integration and Transformation
- Code Asset Management


Main features:


Catalog:
>  A catalog contains metadata about the contents of assets and how to access them. And a set of collaborators who need to use the assets for data analysis

Difference between Using Catalog vs. Uploading data:
- you upload your data to the Cloud Object Storage
- then specify that object storage when you create the catalog.
- please note: Catalog only stores metadata, not the actual your uploaded data
- This split between where the data's metadata is stored and the actual location of the data is important.
- It means that you can keep your data where ever it is. 
    - You don't need to move it into the catalog because the catalog only contains metadata. 
    - You can have the data in unpremises data repositories in other *IBM cloud services* like **Cloudant or Db2 on Cloud** and in *non-IBM cloud services* like **Amazon or Azure, in streaming data services or even dark data sources like PDFs.**

### SPSS Modeler Flows


- SPSS Modeler Flows is a part of Watson Studio, which was inspired by another product, IBM SPSS Modeler
- It can cover the following DS tasks (see green):
    - Data Management
    - Data Integration and Transformation
    - Data Visualization
    - Model Building

- All flows are created using a drag-and-drop editor and consist of nodes of various types, with data flowing from one node to the next according to their connections



Example:
- two data source nodes are shown in purple on the left
- type, aggregate, filter, merge, filler, and partition nodes in the middle
- 2 model building nodes shown in pentagons. 
- Once a flow is executed and the models are built, the upside-down pentagon model nuggets are created. They can be used to see information about the models and to get predictions for new data. 
- And the three green square nodes on the right provide model evaluation information in the form of tables and charts.

Choosing model:
- You can customize which model to use in "Modeling" option
- Please note: the first three models with "auto", will automatic choose the best model for you, in each classification type models, numeric type models, or clustering models.




### SPSS Modeler


- This product was created by Integral Solutions Limited in the United Kingdom in 1994 and was originally called **Clementine**. It was acquired by a company called SPSS in 1998 and SPSS was in turn acquired by IBM in 2009.
- It can cover the following DS tasks (see green):
    - Data Management
    - Data Integration and Transformation
    - Data Visualization
    - Model Building
    - Model Deployment
    
   
   
- SPSS Modeler is a data mining and text analytics software application. 
- It's used to build predictive models and conduct other analytics tasks. 
- It has a visual interface that enables users to leverage statistical and data mining algorithms without programming.
- One of its main goals from the beginning was to create complex predictive modeling pipelines that are easily accessible. 

### SPSS Statistics

- IBM SPSS Statistics evolved from an original product that was released in 1968. That product was called Statistical Package for Social Sciences, or SPSS.
- IBM SPSS Statistics is a statistical and machine learning software application and is widely used in academia, government agencies, and large enterprises.
- Its used to build predictive models, perform statistical analysis of data, and conduct other analytic tasks. 
- It has a visual interface, which enables users to leverage statistical and data mining algorithms without programming, although the interface is very different from Modeler.
    - main section of the screen looks very much **like a spreadsheet**; it displays data and allows manual editing. 
    

- Using a graphical user interface, we can create complicated analysis that can be saved in the form of syntax and reused later.

### Deploying Models

Open standards for model deployment are designed to support model exchange between a wider variety of proprietary and open source models.

#### a. PMML standard: (Predictive Model Markup Language)
- First open standard
- XML language based
- The products we looked at earlier -- Watson Studio, IBM SPSS Statistics, IBM SPSS Modeler -- enable users to export most models in PMML. 

#### b. PFA (Portable Format for Analytics):
- Did not describe models and their features, but rather the scoring procedure directly
- based on JSON

#### c. ONNX (Open Neural Network Exchange)
- initially specific for NN model
- later also can deploy traditional ML model

#### d. IBM Watson ML Deployment method
- Deploy models built with Watson Studio, SPSS Modeler, many open source packages
- Support for models expressed in PMML and ONNX
- Deployment can be done via GUI or Python code
- Scoring can be: online scoring via REST API or batch scoring
