## 2.2. Magic Functions of Matplotlib

### 2.2.1. %matplotlib inline:

- Enforce the plot is rendered/generated within the browser and not in a separate window, by passing inline as backend.

### 2.2.2. %matplotlib notebook

- Enforce the plot is rendered/generated within the browser similar to inline backend above.
- It checks if the active figure exists in the notebook, if not, then generate the new figure same as inline backend above.
- What is different from inline backend above is that:
    - if the active figure exists in the notebook, it will add anything function you called to this existing figure (e.g. add title to the figure);
    - compared to inline backend, in this case, after a figure is .show(), it won't add anything to that existing figure.
