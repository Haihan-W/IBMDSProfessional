## List is mutable

```python
a=[1,2,3]
a[1]=0
a
```

### Because List is Mutable, therefore there is deep copy and shallow copy of the list

- Deep Copy: i.e. reference -> B=A, B and A reference the same list, if elements in A changes, then B got changed as well 
- Shallow Copy: i.e. clone -> B=A[:], B and A reference the two lists, if elements in A changes, then B will not change

```python
#Deep Copy
A=[1,2,3]
B=A
A[0]=3234
B
```

```python
#Shallow Copy
A=[1,2,3]
B=A[:]
A[0]=3234
B
```

List Concatenation

```python
a=[1,2,3]
a.append(4)
a
```

```python
a=[1,2,3]
a.append([4])
a
```
