## 1.2. Multiple Linear Regression (multi "independent" x, 1y)

**How to estimate the coefficients?**

principle: Ordinary Least Square -- i.e. try to minimize error (e.g. MSE)


- Option 1: Mathematical Approach (good for data size < 10K rows)



- Option 2: Optimization Approach (good for large data size)


**How to determine whether I should include an independent variable to MLR model?**
- VIP: Note MLR should ideally have y as linear relationship with each of independent varialble x, so scatter plot that y vs x for the x you want to add to the model: if relationship is non-linear, then you need to consider non-linear regression. (note: for each x, if other independent variables not change, the change in x should be proportional to change in y.)

# 3. Non-Linear Regression

Definition: Relationship between y and **coefficient** is non-linear, instead of relationship bewtween y and independent variable x!!!!!!

So by this definition, polynomial regression does not fit under this category...

Also, for non-linear regression, we cannot use Least Squares method to estimate the coefficients because it is unable to convert to MLR.
