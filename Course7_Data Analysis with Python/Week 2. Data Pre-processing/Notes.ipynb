# 1. Deal with Missing Values

### Methods:
- Check with data collection source

- Drop the missing values (df.dropna())
    - Drop that entire data variable (i.e. the column)
    - Drop specific data entry (i.e. the row)
    
- Replace the missing values (df.replace(np.nan,new_value), df.replace("?",new_value), ...)
    - replace it with an average (of similar datapoints) (i.e. this is for numerical value only)
    - replace it by frequency - this is for categorical value (e.g. the most common categorical value, It is called "Mode")
        - ***code example***: 
            - most_common = df['num-of-doors'].value_counts().idxmax() -> returns the most common categorical value in the col;
            - df["num-of-doors"].replace(np.nan, most_common , inplace=True) -> replace nan with the most common categorical value.
    - replace it based on other functions
    
- Leave as missing data
