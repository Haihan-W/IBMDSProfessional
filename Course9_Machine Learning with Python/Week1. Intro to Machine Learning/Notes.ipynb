## Major ML techniques
1. Regression/Estimation:
    - Predicting continuous values
    - Example: car CO2 emission, house price estimation

2. Classification:
    - Predicting the item's class/category of a case
    - Example: If a cell is cancer type cell or not.
    
3. Clustering:
    - Finding structure of data and summarization
    - Example: Customer segmentation in banking
    
4. Associations:
    - Associating frequent co-occurring items/events
    - Example: Finding the grocery items that are usually bought together by customer
    
    
5. Anomaly detection: (one application of **clustering** techniques)
    - Discovering abnormal and unusual cases
    - Example: Credit card fraud detection
    
6. Sequence Mining:
    - Predicting next events
    - Example: click-stream (Markov Model, HMM) - i.e. a user's activities on the World Wide Web as represented by the sequence of links they click on, this is called click-stream
    
7. Dimension Reduction:
    - Reducing size of data (PCA)
    
8. Recommendation Systems:
    - Recommending items to specific user.
    - Amazon recommend items to user based on purchasing history of other users with similar characteristics.

## Supervised vs Unsupervised Learning

1. Supervised Learning (2 types of techniques)
    - 
    - Classification -> predict discrete Y
    - Regression -> predict continuous Y
    
2. Unsupervised Learning (common types of techniques below)
    - Clustering
    - Dimension reduction/feature selection
    - Density Estimation
    - Market basket analysis (like "Association" technique mentioned above)
