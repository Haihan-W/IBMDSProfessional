Based on the videos and the reading material, how would you define a data scientist and data science? (3 marks)

Data science is the study of large quantities of data, which can reveal insights that help organizations make strategic choices.
Data scientist uncovers the insights and trends that are hiding behind data, and translates data into a story using storytelling, and generates insight to help organization to make data-driven decisions.
